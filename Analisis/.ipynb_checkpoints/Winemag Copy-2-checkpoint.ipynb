# Preprocesamiento

```python
#Para poder realizar la conexión entre variables climáticas y calificación ed los vinos, requqrimos tener una
#Llave de conexión entre los dataframes. Esta llave es la fecha de producción del vino, la cual se toma del
#título del vino. Para esto se manipula la columna 'title', donde a través de regex, se extraen
#los valores de años 2000 a 2099 (solo llegan hasta 2017), estos valores se guardan en una nueva columna
#llamada 'Year', se reemplazan los valores nulos por cero y se convierte la columna a tipo entero
data['Year'] = data['title'].str.extract('([2][0][0-9][0-9])', expand=False).str.strip()
data['Year'] = data['Year'].fillna(0) #cambiar valores nulos a ceros
data['Year'] = data['Year'].astype(int)
```

```python
#Se imprimen solo los valores únicos, sin encontrar datos atípicos. El valor '0' son las entradas que no se 
#encontraron años de producción de los vinos
data['Year'].unique()
```

```python
#Se eliminan los datos de año '0'. Además se requieren eliminar los datos desde el 2013 en adelante, ya que
#los dataframes de clima que se encontraron solo tienen datos hasta 2012 inclusive
data.drop(data[(data['Year']>= 2013)|(data['Year'] == 0)].index, inplace = True)
#Nuevamente se validan los valores de datos únicos, los cuales solo corresponen de 2000 a 2012
data['Year'].unique()
```

```python
#Se instancia el objeto geolocator, usando en este caso la API de Google con su respectiva APIKey (se remueve por 
#temas de costos y procesamiento público).
#Así mismo, luego de correr el código la primera vez (timepo computacional approx 750 segundos), se recomienda 
#comentar este código ya que los resultados son guardados en un archivo CSV para su uso en otras fases del análisis

geolocator = geocoders.GoogleV3(api_key="")
```

```python
#Se crea una funcion para la evaluación de la geolocalización, en el cual se le pasa al geocoder el dato de la region
#por ejemplo, "Douro, Portugal", y este devuelve las coordenadas de Latitud y Longitud según su base de datos
#Para esto se incluye un try/except, ya que en caso que el sistema de geolocalización no obtenga información de este 
#sitio, se debe #agregar el dato "None", en Latitud/Longitud para su respectivo descarte.

def eval_results(x):
    try:
        return (x.latitude, x.longitude)    
    except:
        return (None, None)
```

```python
%%time
#Se usa un dataset temporal (data_temporal), donde se aplica el la evaluación de la geolocalización por cada una de
#las entradas. En 'data_temporal' se agregan en la columna 'region_coord' los valores de Latitud y Longitud de cada
#entrada de región (entrada única)
#Tambien se analiza el tiempo computacional en realizar esta validación cruzada en API (750 segundos aprox.)
data_temporal = data_region_unique_values
data_temporal['region_coord'] = data_temporal['region'].apply(geolocator.geocode).apply(lambda x: eval_results(x))
data_temporal
```

```python
#Luego de que el código termina, se guarda en un archivo CSVla información obtenida de Google, para así ahorrar 
#tiempo computacional en el análisis posterior. Así mismo se sugiere cometnar las líneas del bloque anterior y actual
#Luego de ser procesadas por primera vez
data_temporal.to_csv('data_temporal_locations.csv', index = True)
```

```python
#La columna 'region_coord' contiene los valores de Latitud y Lonfgitrud en forma de objeto, por lo cual se debe
#procesar para separar la información de Latitud y Longitud y a continuación transformar esta el tipo numérica (float64)
#En la nueva columna 'coord' se insertan los valores de 'region_coord' eliminando los caracteres de parentesis y comas 
#que no brindan información, a través de código RegEx

dataframe.insert(2,'coord', dataframe['region_coord'].str.replace('(\()|(,)|(\))','', regex=True))
#Se visualiza el dataset actual
dataframe.head()
```

```python
#Se convierten las nuevas columnas en tipo float64 para conservar los decimales. Finalmente se eliminan las columnas
#'region_coord' y 'coord' ya que no están aportando información adicional al dataset

dataframe['Latitude'] = dataframe.Latitude.astype(np.float64)
dataframe['Longitude'] = dataframe.Longitude.astype(np.float64)
dataframe.drop(columns = 'region_coord', inplace = True)
dataframe.drop(columns = 'coord', inplace = True)
#Se valida que las columnas Latitude y Longitude queden como tipo float64 (numérico)
dataframe.info()
```

## Unión Dataset final

```python
#Se obtiene el dataframe de los datos de clima (dataclima), y se visualizan para analizar su estructura 

dataclima = pd.read_csv('Data_clima_final.csv')
dataclima.head()
```

```python
#Se obtienen los datos ed año (agno), temperatura anual promedio (temp_anual), termperatura máxima y mínima 
#anual (temp_max_anual, temp_min_anual), precipitación anual (pre_anual), evotranspiración anual (etp_anual) y 
#las coordenadas de cada punto de análisis
#Se procede a limpiar al columna 'coordenadas', eliminando los caracteres de parentesis y coma que no proporcionan
#información útil. Esta información se almacena en la nueva columna 'coord'

dataclima.insert(2,'coord', dataclima['coordenadas'].str.replace('(\()|(,)|(\))','', regex=True))
#Se visualiza la información actual del dataset
dataclima.head()
```

```python
#Al igual que el dataset de winemag, se realiza el mismo preprocesamiento con las coordenadas, donde se crean
#nuevas columnas de Latitude y Longitude a través de un split, y se les asigna el tipo de dato float64

dataclima[['Latitude','Longitude']] = dataclima.coord.str.split(expand=True) 
dataclima['Latitude'] = dataclima['Latitude'].astype(np.float64)
dataclima['Longitude'] = dataclima['Longitude'].astype(np.float64)
#Se visualiza este dataset para validar el cambio de tipo de dato
dataclima.info() 
```

```python
#Se procede a eliminar las columnas 'coordenadas', 'coord' y 'Unamed: 0' ya que no aportan información adicional

dataclima.drop(columns = 'coordenadas', inplace = True)
dataclima.drop(columns = 'coord', inplace = True)
dataclima.drop(columns = 'Unnamed: 0', inplace = True)
#Se visualiza el último dataset
dataclima.head()
```

```python
#Para realizar la conexión de datos entre el dataset de clima y winemag, se requiere que se analicen cada punto (Lat/Long)
#del dataset winemag y se crucen con los puntos ed Lat/Long del dataset de clima, y uscar el punto más cercano,
#para así obtener la información de la estación de clima más cercano al sitio de la región donde fue producido el vino
#Para esto, Se crean dos funciones, evaluando la ocurrencia del valor más cercano y retornando este valor en una 
#variable (Lat o Long). De encontrar un error en este análisis, retorna un valor "NaN"

def eval_latitude(x, df):
    try:
        Lat = df.iloc[(df.Latitude-x).abs().argsort()[:1]].Latitude.values[0]
        return Lat
    except:
        return "NaN"
    
def eval_longitude(x, df):
    try:
        Long = df.iloc[(df.Longitude-x).abs().argsort()[:1]].Longitude.values[0]
        return Long
    except:
        return "NaN"
```

```python
%%time
#Se procede a realizar el análisis de cada valor en 'Latitude' y 'Longitude' del dataset Winemag contra 
#el dataset de clima, almacenando el valor más cercano en las nuevas columnas 'Lat_x' y 'Long_x'
#También se analiza el tiempo total de este análisis. Se sabe de antemano que se puede bajar este tiempo computacional
#corriendo el código solamente en el dataset de regiones únicas, esto puede ser una optimización de código a futuro
#Se recomienda comentar este análisis a futuro para evitar tiempo computacional

dataframe['Lat_x'] = dataframe['Latitude'].apply(lambda x: eval_latitude(x, dataclima))
dataframe['Long_x'] = dataframe['Longitude'].apply(lambda x: eval_longitude(x, dataclima))
```

```python
#Se visualiza en último dataset Winemag para encontrar la conexión de valores de Latitud y Longitud
#entre ambos datasets (Lat_x y Long_x)
dataframe.head(2)
```

```python
#Se valida que las nuevas columnas tengan tipo de dato float64
dataframe.info()
```

```python
#Se guarda la información de este dataset en un archivo CSV para ahorrar tiempo computacional.
dataframe.to_csv('Dataframe_winereviews_.csv', index = True)
```

```python
#En el dataset de clima, se renombran las variables de Latitude y Longitude por Lat_x y Long_x para que coincidan
#con los nombres en el dataset winemag, así como el año (agno:Year)

dataclima = dataclima.rename(columns={'Latitude':'Lat_x'
                                      , 'Longitude':'Long_x'
                                      ,'agno':'Year'})
#Se visualiza el último dataset de clima
dataclima.head()
```

```python
#En datapro se realiza el "merge" o unión de los dos datasets, realizando esta unión con las variables de año,
#Lat_x y Long_x. Es así que finalmente se obtiene el último dataset con la unión de parámetros
datapro = pd.merge(dataframe,dataclima,on=['Lat_x','Long_x', 'Year'])
datapro.head()
```

```python
# Se valida el tipo ed dato de cada columna 

datapro.info()
```

```python
#Se guarda este dataset final en un CSV para ahorra costo y tiempo computacional

datapro.to_csv('Wine_reviews_climate_prediction.csv', index = True)
```
